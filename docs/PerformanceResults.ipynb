```python
import twitter_location
import pandas as pd
```

```python
results = pd.read_excel("../data/test_data.xlsx")
results = results[results["Original"].notna()]
```

```python
results
```

```python
predictions = []
for i in results.index:
    location = results.loc[i]["location"]
    lang = results.loc[i]["lang"]
    predictions.append(twitter_location.prediction.predict(location=location,language=lang))
results["prediction"] = predictions
```

```python
def CorrectnessTable(df):
    orig = df["Original"]
    data = {}
    col = "prediction"
    correct = sum(orig == df[col])
    incorrect = sum((orig != df[col]))
    total = correct+incorrect
    ratio = float(correct)/total
    data[col] = [correct,incorrect,total,ratio]
    return pd.DataFrame(data=data,index = ["Correct","Incorrect","Total","Ratio"])
```

```python
CorrectnessTable(results)
```
